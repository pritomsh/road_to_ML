```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
df=pd.read_csv('data.csv')
```

```python
df
```

```python
x = df['test']
y = df['class']
```

```python
x.head()
```

```python
cv=CountVectorizer()
```

```python
new_x = cv.fit_transform(x)
```

```python
new_x
```

```python
new_x.toarray()
```

```python
cv.get_feature_names()
```

```python
freq = pd.DataFrame(new_x.toarray(), index = df['test'], columns=cv.get_feature_names())
```

```python
freq
```
